```python
from sklearn.impute import KNNImputer
knnimp = KNNImputer(n_neighbors=3)
```

```python
cols = ['age','TSH','T3','TT4','T4U','FTI']
for i in cols:
    df[i] = knnimp.fit_transform(df[[i]])
```

```python
df.isnull().sum()
```

```python
#some of the features are in categorial object so we use label encoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
variables = ['age', 'sex', 'on thyroxine', 'query on thyroxine', 'on antithyroid medication:', 
             'sick', 'pregnant', 'thyroid surgery', 'I131 treatment', 'query hypothyroid', 
             'query hyperthyroid', 'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych', 
             'TSH measured', 'TSH', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4u measured', 
             'T4U', 'FTI measured', 'FTI', 'TBG measured', 'TBG']
```

```python
# Update dataset with label encoded features inplace
for variable in variables:
    if variable in df.columns:
        df[variable] = le.fit_transform(df[variable])
```

```python
# now we can see their is correlation in some features
for a in range(len(df.corr())):
    for b in range(a):
        if((df.corr().iloc[a,b]) >= 0.7):
            print(df.corr().columns[b])
```

```python
# so TT4 has high correlation among all we can remove

df.drop('TT4',axis=1,inplace=True)
```

```python
X = df.drop('Output',axis=1)
y = df.Output
df2 = X
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=10000)
lrclf = lr.fit(X_train,y_train)
y_pred = lrclf.predict(X_test)
accuracy_score(y_pred,y_test)
```
